## 정리노트(210104)

### 아나콘다 설치 및 사용자 설정

- Conda update

  ```bash
  conda update --all
  ```

- PIP Update

  ```bash
  python -m pip install --upgrade pip
  ```

- Check

  ```bash
  conda --version
  ```

### 가상환경 생성 및 설정

- Create virtual env & install python 3.7 version, openssl

  ```bash
  conda create -n [name] python=3.7 openssl
  ```

- Check

  ```bash
  conda env list
  conda info --envs
  ```

- Activate

  ```bash
  conda activate data_env
  ```

### Jupyter notebook 

- Install notebook

  ```bash
  conda install nb_conda
  ```

- Setting Home directory

  ```bash 
  jupyter notebook --generate-config
  ```
  해당 경로의 파일을 열고 385번째 줄을 주석해제 후 원하는 저장 경로 설정

  ```bash
  c.NotebookApp.notebook_dir = '저장경로'
  ```

- 
  Execution
  
    ```bash
    jupyter notebook
    ```

- Cell
  
    - 위쪽에 cell 추가 => [ESC], [A]
    - 아래쪽에 cell 추가 => [ESC], [B] 
    - cell 제거 => [ESC], [D] [D]
    - markdown 진입 => [ESC], [M]
    - code 진입 => [ESC], [Y]
    - code 실행 => [Ctrl] + [Enter]
